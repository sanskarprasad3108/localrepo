```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("dump_truck_industrial_anomaly_dataset_1200.csv")
df.head()
```

```python
features = [
    "engine_temp_C",
    "oil_pressure_bar",
    "hydraulic_pressure_bar",
    "fuel_rate_lph",
    "vibration_mm_s",
    "brake_temp_C",
    "speed_kmph",
    "load_tons"
]

X = df[features]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
iso_forest = IsolationForest(
    n_estimators=200,
    contamination=0.15,   # approx anomaly percentage
    random_state=42
)

df["anomaly_score"] = iso_forest.fit_predict(X_scaled)
```

```python
df["predicted_label"] = df["anomaly_score"].map({1: 0, -1: 1})
df[["label", "predicted_label"]].head()

```

```python
pd.crosstab(df["label"], df["predicted_label"],
            rownames=["Actual"], colnames=["Predicted"])
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x=df["engine_temp_C"],
    y=df["vibration_mm_s"],
    hue=df["predicted_label"],
    palette={0: "blue", 1: "red"}
)

plt.xlabel("Engine Temperature (C)")
plt.ylabel("Vibration (mm/s)")
plt.title("Dump Truck Anomaly Detection (Isolation Forest)")
plt.legend(title="Anomaly (1 = Yes)")
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
accuracy = accuracy_score(df["label"], df["predicted_label"])
precision = precision_score(df["label"], df["predicted_label"])
recall = recall_score(df["label"], df["predicted_label"])
f1 = f1_score(df["label"], df["predicted_label"])

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1-Score :", f1)
```

```python
print(classification_report(df["label"], df["predicted_label"]))
```
